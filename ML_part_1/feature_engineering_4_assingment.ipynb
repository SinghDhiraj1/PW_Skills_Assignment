# Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

```python
"""
Nominal encoding perform for the data which cannot be ranked and it converts the data into binary vector representation for all categories.
"""
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

df = sns.load_dataset('tips')
df.head()
```

```python
### df has a column 'day' which can be converted to numeric using nominal encoding using ohe(one hot endoder)
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['day']])
day_df = pd.DataFrame(encoded.toarray(), columns = encoder.get_feature_names_out())
pd.concat([day_df, df], axis = 1).head()
```

# Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.

```python
"""
nominal encoding and one-hot encoding are the same technique for preprocesing the categorical data to convert it into numeric

nominal or one-hot encoding is used where:
1. **No ordinal relationship exists:** One-hot encoding is preferred when there is no meaningful order among categorical values.
2. **Avoiding model bias:** One-hot encoding prevents the model from assuming a false ordinal relationship between categories.
3. **Handling nominal variables with multiple categories:** One-hot encoding is effective for nominal variables with numerous categories.
"""
df1 = sns.load_dataset('diamonds')
df1.head()
```

```python
"""
In the above dataframe color category do not have any relationship we can use one-hot or nominal encoding
"""
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df1[['color']])
color_df = pd.DataFrame(encoded.toarray(), columns=encoder.get_feature_names_out())
pd.concat([df1, color_df], axis = 1).head()
```
