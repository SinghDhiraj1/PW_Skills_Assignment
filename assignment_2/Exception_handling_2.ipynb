```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
"""
Errors that are defined in the ArithmeticError are 
1. ZeroDivisionError
2. OverflowError
3. FloatingPointError
"""

## ZeroDivisionError occured when we divide any number with Zero
try:
    print(5/0)
except ZeroDivisionError:
    print("You have divided with zero")
```

```python
## OverFlowError is occured when we Exceed the memory space for the Integer
import math
try:
    print("The exponential value is")
    print(math.exp(1000))
    
except OverflowError as oe:
    print("After overflow", oe)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
"""
It occured when we access the Index that is not present in the sequence or iterators
"""
try:
    list1 = [1,2,3,4]
    print(list1[4])
except LookupError:
    print("Index out of bound")
```
