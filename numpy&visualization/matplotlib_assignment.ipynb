```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
"""
scatter plot are the plotting of the data points(bot) from x and y axis which represents the plot in the graph plane
"""
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
colour = np.random.rand(50)
sizes = 500*np.random.rand(50)

plt.scatter(x,y, c = colour, s = sizes, alpha=0.7)
plt.xlabel("This is xlabel")
plt.ylabel("This is ylable")
plt.title("Scatter Plot Visualiztion")
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

"""
subplot function is used for plot muliple plots in a single plane to save time and get more insites from data quickly

"""

# import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)

plt.suptitle('subplot')
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

"""
bar plot is representation of plot between a categorical and numerical data to show data insites
"""

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```

```python
plt.bar(company, profit)
plt.xlabel("company name")
plt.ylabel("profit ($)")
plt.title("Question no. 4 plot")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel("company name")
plt.ylabel("profit ($)")
plt.title("Question no. 4 plot")
plt.show()
```
